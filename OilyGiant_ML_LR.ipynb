# Entrenamiento del modelo

```python
# división del df0 en características y objetivo
features = df0[['f0', 'f1', 'f2']]
target = df0['product']

# división del conjunto de datos en entrenamiento y validación
features_train, features_valid, target_train, target_valid = train_test_split(features, target, test_size=0.25, random_state=42)
```

```python
# creación y entrenamiento del modelo
model = LinearRegression() # condición mencionada
model.fit(features_train, target_train)

# validación del modelo
target_pred = model.predict(features_valid)
```

```python
# guardado de las predicciones y respuestas correctas
df_predictions = pd.DataFrame({'predictions': target_pred, 'answers': target_valid})
print(df_predictions.head())
```

```python
# cálculo del volumen medio de reservas
average_vol_reserves = df_predictions['predictions'].mean()
print(f"Volumen medio de las reservas predicho: {average_vol_reserves:.4f} miles de barriles")

# cálculo del RMSE
rmse = mean_squared_error(df_predictions['answers'], df_predictions['predictions']) ** 0.5

print(f"Raíz del error cuadrático medio (RMSE): {rmse:.4f} miles de barriles")
```

## Resultados
El volumen medio de las reservas predicho indica cuánto se espera que contenga un pozo promedio, mediante regresión lineal

El RMSE muestra el error promedio de las predicciones con respecto a las reservas reales. Cuanto más bajo sea el RMSE, mejor será el rendimiento del modelo

# Creación de funciones

```python
def entrenar_evaluar_modelo(data):
    # separación en características y objetivo
    features = data[['f0', 'f1', 'f2']]
    target = data['product']

    # división en conjunto de datos de entrenamiento y validación
    features_train, features_valid, target_train, target_valid = train_test_split(features, target, test_size=0.25, random_state=42)

    # creación y entrenamiento del modelo
    model = LinearRegression()
    model.fit(features_train, target_train)

    # predicción del modelo
    target_pred = model.predict(features_valid)

    # cálculo del volumen medio de reservas y RMSE
    average_vol_reserves = np.mean(target_pred)
    rmse = mean_squared_error(target_valid, target_pred) ** 0.5

    return target_pred, target_valid, average_vol_reserves, rmse   
```

```python
# aplicación de la función creada en todos los df

target_pred_0, target_valid_0, average_vol_0, rmse_0 = entrenar_evaluar_modelo(df0)
print(f"Región 0 - volumen medio de reservas {average_vol_0:.4f} miles de barriles, RMSE: {rmse_0:.4f} miles de barriles")

target_pred_1, target_valid_1, average_vol_1, rmse_1 = entrenar_evaluar_modelo(df1)
print(f"Región 1 - volumen medio de reservas {average_vol_1:.4f} miles de barriles, RMSE: {rmse_1:.4f} miles de barriles")

target_pred_2, target_valid_2, average_vol_2, rmse_2 = entrenar_evaluar_modelo(df2)
print(f"Región 2 - volumen medio de reservas {average_vol_2:.4f} miles de barriles, RMSE: {rmse_2:.4f} miles de barriles")
```

## Función para calcular la ganancia de un conjunto de pozos seleccionados

```python
def calcular_ganancia_pozos_top(predictions):
    # valores predefinidos
    wells = 200
    price_per_unit=4500
    
    # ordenar las predicciones de mayor a menor
    sorted_predictions = np.sort(predictions)[::-1]

    # seleccionar las predicciones más altas (200 pozos)
    top_wells = sorted_predictions[:wells]

    # calcular el volumen total de reserva
    total_volume = top_wells.sum()

    # calcular la ganancia total
    total_profit = total_volume * price_per_unit

    return total_volume, total_profit   
```

```python
# evaluación de la función
total_volume_0, total_profit_0 = calcular_ganancia_pozos_top(target_pred_0)
print(f"Región 0 - volumen total de reservas de los mejores 200 pozos: {total_volume_0:.4f} miles de barriles")
print(f"Región 0 - Ganancia total potencial: ${total_profit_0:.2f} USD")

total_volume_1, total_profit_1 = calcular_ganancia_pozos_top(target_pred_1)
print(f"\nRegión 1 - volumen total de reservas de los mejores 200 pozos: {total_volume_1:.4f} miles de barriles")
print(f"Región 1 - Ganancia total potencial: ${total_profit_1:.2f} USD")

total_volume_2, total_profit_2 = calcular_ganancia_pozos_top(target_pred_2)
print(f"\nRegión 2 - volumen total de reservas de los mejores 200 pozos: {total_volume_2:.4f} miles de barriles")
print(f"Región 2 - Ganancia total potencial: ${total_profit_2:.2f} USD")
```

# Cálculo de los riesgos para cada región

```python
def bootstrapping(predictions, n_samples=1000):
    state = np.random.RandomState(42)
    profits = []
    predictions = pd.Series(predictions)

    for i in range(n_samples):
        pred_sample = predictions.sample(n=200, replace=True, random_state=state)
                
        # ganancia total de la muestra
        profit = sum(pred_sample) * 4500 
        profits.append(profit)

    profits = pd.Series(profits)

    # cálculo del beneficio promedio promedio
    mean_profit = profits.mean()

    # cálculo del intervalo de confianza del 95%
    confidence_interval = st.t.interval(confidence=0.95, df=len(profits)-1, loc=profits.mean(), scale=profits.sem())

    # cálculo del riesgo de pérdidas
    lost_risk = np.mean(profits < 0) 

    return mean_profit, confidence_interval, lost_risk
```

```python
profit_0, interval_0, risk_0 = bootstrapping(target_pred_0)
print(f"Región 0 - Beneficio promedio: ${profit_0:.4f}, Intervalo de confianza 95%: {interval_0}, Riesgo de pérdida: {risk_0:.1%}")

profit_1, interval_1, risk_1 = bootstrapping(target_pred_1)
print(f"\nRegión 1 - Beneficio promedio: ${profit_1:.4f}, Intervalo de confianza 95%: {interval_1}, Riesgo de pérdida: {risk_1:.1%}")

profit_2, interval_2, risk_2 = bootstrapping(target_pred_2)
print(f"\nRegión 2 - Beneficio promedio: ${profit_2:.4f}, Intervalo de confianza 95%: {interval_2}, Riesgo de pérdida: {risk_2:.1%}")
```
